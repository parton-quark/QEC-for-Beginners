```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit import execute
from qiskit import IBMQ, Aer
backend = Aer.get_backend('statevector_simulator')

import numpy as np
```

# ideal

```python
q = QuantumRegister(1)
c = ClassicalRegister(1)
qc = QuantumCircuit(q,c)
theta = np.pi
# qc.rx(theta, 0)
qc.i(0)
qc.measure(q,c)
qc.draw(output="mpl")
```

```python
job = execute(qc, backend)
result = job.result()
outputstate = result.get_statevector(qc, decimals=3)
print(outputstate)
```

```python
from qiskit.visualization import plot_state_city
plot_state_city(outputstate)
```

# Coherent error

```python
qch = QuantumRegister(1)
cch = ClassicalRegister(1)
qcch = QuantumCircuit(qch,cch)
theta = np.pi
qcch.rx(theta, 0)
# qcch.i(0)
qcch.measure(qch,cch)
qcch.draw(output="mpl")
```

```python
jobch = execute(qcch, backend)
resultch = jobch.result()
outputstatech = resultch.get_statevector(qcch, decimals=3)
print(outputstatech)
```

```python
plot_state_city(outputstatech)
```
